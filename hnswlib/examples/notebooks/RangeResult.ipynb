```python
import pandas as pd
import glob
import os

# Path where all your CSV files are stored (change this)
path = "/data4/hnsw/TripClick/Ground_truth_EDBT2026_complete/"   # <-- folder containing 1.csv, 2.csv, ...

# Get list of all CSV files
files = glob.glob(os.path.join(path, "*.csv"))

results = []

for file in files:
    # Extract file name without extension → this becomes the id
    file_id = os.path.splitext(os.path.basename(file))[0]
    
    # Read the CSV
    df = pd.read_csv(file)
    
    # Compute average distance
    avg_distance = df['Distance'].mean()
    
    # Save to results list
    results.append([file_id, avg_distance])

# Convert results to DataFrame
output_df = pd.DataFrame(results, columns=["id", "average_distance"])

# Save final result
# output_df.to_csv("final_avg_distances.csv", index=False)

# output_df.head()
```

```python

# Round the average_distance to 3 decimal places
output_df['average_distance'] = output_df['average_distance'].round(3)
```

```python
# Extract numeric part from id (works for Q0, Q12, 5, etc.)
output_df['id_num'] = output_df['id'].str.extract('(\d+)').astype(int)

# Sort by numeric value
output_df = output_df.sort_values('id_num').drop(columns=['id_num'])
```
